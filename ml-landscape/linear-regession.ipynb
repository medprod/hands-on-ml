```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
```

### Part 2

```python
plot.scatter(lindf.X, lindf.Y, color='gold')
```

### Part 4

#### K = 4:

```python
K = 4
nonlin_x = np.array(nonlindf.X)
nonlin_y = np.array(nonlindf.Y)

# print(nonlin_x)
xfunc_vals = np.linspace(0,15,10000)
yfunc_vals = []

for xval in xfunc_vals:
    distances = np.abs(xval - nonlin_x)
    sorted_index = np.argsort(distances)[:K]
    
    yvals = []
    for i in sorted_index:
        yvals.append(nonlin_y[i])
    
    res = np.mean(yvals)
    yfunc_vals.append(res)
    
plot.scatter(nonlin_x, nonlin_y, color='gold')
plot.plot(xfunc_vals, yfunc_vals, color='red')
```

### Part 5

#### Function Plot

```python
N = len(lindf.X)

xavg = np.mean(lindf.X)
yavg = np.mean(lindf.Y)
xsq_avg = np.mean(lindf.X * lindf.X)
ysq_avg = np.mean(lindf.Y * lindf.Y)
xy_avg = np.mean(lindf.X * lindf.Y)

def fun(b0, b1):
    h = N * (ysq_avg + b1**2 * xsq_avg + 2 * b0 * b1 * xavg - 2 * b0 * yavg - 2 * b1 * xy_avg + b0**2)
    return h

b0_vals = np.arange(-100, 100, 0.1)
b1_vals = np.arange(-100, 100, 0.1)

B0, B1 = np.meshgrid(b0_vals, b1_vals)
H = fun(B0, B1)

fig = plot.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(B0, B1, H, color='green')

ax.set_xlabel('B0')
ax.set_ylabel('B1')
ax.set_zlabel('H')

plot.show()
```

#### Minimum Point

```python
N = len(lindf.X)

xavg = np.mean(lindf.X)
yavg = np.mean(lindf.Y)
xsq_avg = np.mean(lindf.X * lindf.X)
ysq_avg = np.mean(lindf.Y * lindf.Y)
xy_avg = np.mean(lindf.X * lindf.Y)

b0 = (yavg*xsq_avg - xavg*xy_avg)/(xsq_avg - xavg**2)
b1 = (xy_avg - xavg*yavg)/(xsq_avg - xavg**2)

def fun(b0, b1):
    h = N * (ysq_avg + b1**2 * xsq_avg + 2 * b0 * b1 * xavg - 2 * b0 * yavg - 2 * b1 * xy_avg + b0**2)
    return h

b0_vals = np.arange(-100, 100, 0.1)
b1_vals = np.arange(-100, 100, 0.1)

B0, B1 = np.meshgrid(b0_vals, b1_vals)
H = fun(B0, B1)

fig = plot.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(B0, B1, H, color='green')
ax.scatter(b0, b1, fun(b0, b1), color='red', s=100, label='Minimum Point')

ax.set_xlabel('B0')
ax.set_ylabel('B1')
ax.set_zlabel('H')
ax.view_init(8,20)

ax.legend()
plot.show()
```

#### Function Plot

```python
def g(b0, b1):
    return b0**2 - b1**2

b0_vals = np.arange(-100, 100, 0.1)
b1_vals = np.arange(-100, 100, 0.1)

B0, B1 = np.meshgrid(b0_vals, b1_vals)
G = g(B0, B1)

fig = plot.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(B0, B1, G, color='green')

ax.set_xlabel('B0')
ax.set_ylabel('B1')
ax.set_zlabel('G')

plot.show()
```

#### Saddle Point

```python
def g(b0, b1):
    return b0**2 - b1**2

b0_vals = np.arange(-100, 100, 0.1)
b1_vals = np.arange(-100, 100, 0.1)

B0, B1 = np.meshgrid(b0_vals, b1_vals)
G = g(B0, B1)

fig = plot.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(B0, B1, G, color='green')
ax.scatter(0,0, color='red', s=100, label='Saddle Point')

ax.set_xlabel('B0')
ax.set_ylabel('B1')
ax.set_zlabel('G')

ax.legend()
plot.show()
```
